```python
def CountOccurrences(string, substring): 
    # Initialize count and start to 0 
    count = 0
    start = 0
  
    # Search through the string till 
    # we reach the end of it 
    while start < len(string): 
  
        # Check if a substring is present from 
        # 'start' position till the end 
        flag = string.find(substring, start) 
  
        if flag != -1: 
            # If a substring is present, move 'start' to 
            # the next position from start of the substring 
            start = flag + 1
  
            # Increment the count 
            count += 1
        else: 
            # If no further substring is present 
            # return the value of count 
            return count 

string = "AAGCAACAAGCAACCACCTACAACCAAGCAACACTCAAAGCAACATGGAAGCAACGGGTAAGCAACTTCCGAACCGAAAGCAACAAGCAACCAAAGAAAAGCAACTCACAGCTTACTAGAGGAAAGCAACGAAGCAACCAAAGCAACAAGCAACCGATGAAAGCAACCAAGCAACAAGCAACGAAAGCAACTAAGCAACAAAGCAACAGCCTGGGCTAAAGCAACATGCAAGCAACTAGAAGCAACAAAAGCAACAAGCAACTTTCAAGCAACATCTAAGCAACTGAAGCAACTATTCAAGCAACGAAGCAACTCGGTAAAGCAACGAAGCAACAAGCAACTTGTCGAAGCAACAAGCAACCGAAGCAACTAAGCAACTAAAGCAACCCTAAACTCGAAGCAACACTAGACCAAGCAACAAGCAACCAAGCAACCTAAGCAACAAGCAACAAAGCAACGGGAGTGAAAGCAACGCAAGCAACCTCCTAAGCAACGGAAGCAACGCAAAGCAACAAAGCAACGAAGCAACCTCTGCCCGAAAGCAACCCTGGAAGCAACAAGCAACGGAAAGCAACCCCAAGCAACCCTGAAAGCAACTTGTACTAAGCAACAAGCAACGGCGGAAGCAACATGAAGCAACATAAGCAACACGGCAGAGAAGCAACAAGCAACTAAGCAACAAGCAACTCTAAAGCAACCATAAAAGCAACTTAAGATTCGGGCTTAAGCAACACAAGCAACAAAGCAACAAGCAACAAGCAACAAGCAACCCAAAGCAACAGCGTAAAGCAACCAAGCAACAAGCAACAAGCAACGTCTTCAAGCAACTCAAAGCAACCAAGCAACGGGAACTGAAACGTTTGACGAAGCAACCTACAAGCAACGGCAAAGCAACACAAAGCAACCACTTCCGAAGCAACCTAAGCAACAAGCAACAAGCAACTCTTAAGCAACAAGCAACAAAAGCAACAAGCAACAAGCAACTCGGTATAGAGAAGCAACATAAGCAAC"
print(CountOccurrences(string, "AAGCAACAA"))
```

```python
def count_kmers(read, k, m):
    """Count kmer occurrences in a given read.

    Parameters
    ----------
    read : string
        A single DNA sequence.
    k : int
        The value of k for which to count kmers.

    Returns
    -------
    counts : dictionary, {'string': int}
        A dictionary of counts keyed by their individual kmers (strings
        of length k).

    Examples
    --------
    >>> count_kmers("GATGAT", 3)
    {'ATG': 1, 'GAT': 2, 'TGA': 1}
    """
    # Start with an empty dictionary
    counts = {}
    # Calculate how many kmers of length k there are
    num_kmers = m - k + 1
    # Loop over the kmer start positions
    for i in range(num_kmers):
        # Slice the string to get the kmer
        kmer = read[i:i+k]
        # Add the kmer to the dictionary if it's not there
        if kmer not in counts:
            counts[kmer] = 0
        # Increment the count for this kmer
        counts[kmer] += 1
    # Return the final counts
    return counts

m = 'CACAGTAGGCGCCGGCACACACAGCCCCGGGCCCCGGGCCGCCCCGGGCCGGCGGCCGCCGGCGCCGGCACACCGGCACAGCCGTACCGGCACAGTAGTACCGGCCGGCCGGCACACCGGCACACCGGGTACACACCGGGGCGCACACACAGGCGGGCGCCGGGCCCCGGGCCGTACCGGGCCGCCGGCGGCCCACAGGCGCCGGCACAGTACCGGCACACACAGTAGCCCACACACAGGCGGGCGGTAGCCGGCGCACACACACACAGTAGGCGCACAGCCGCCCACACACACCGGCCGGCCGGCACAGGCGGGCGGGCGCACACACACCGGCACAGTAGTAGGCGGCCGGCGCACAGCC'
k = count_kmers(m,10,len(m))
print(k, len(m))
k1 = [k2 for k2, value in k.items()]
print(k1)
```
